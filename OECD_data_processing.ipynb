```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
CLI_df = pd.read_csv('./data/oecd_data/CLI.csv')[['INDICATOR', 'TIME', 'Value']]
BCI_df = pd.read_csv('./data/oecd_data/BCI.csv')[['INDICATOR', 'TIME', 'Value']]
CCI_df = pd.read_csv('./data/oecd_data/CCI.csv')[['INDICATOR', 'TIME', 'Value']]
```

```python
CLI_df
```

```python
src_df = pd.read_csv('./data/source_price.csv')
src_df
```

```python
for index, row in src_df.iterrows():
    date = row['date'].rsplit('/', 1)[0]
    src_df.at[index, 'CLI'] = CLI_df[CLI_df['TIME'] == date]['Value'].item()-100
    src_df.at[index, 'BCI'] = BCI_df[BCI_df['TIME'] == date]['Value'].item()-100
    src_df.at[index, 'CCI'] = CCI_df[CCI_df['TIME'] == date]['Value'].item()-100
```

```python
src_df
```

```python
src_df.to_csv('./data/source_price_sentiment_oecd.csv', index=False)
```
